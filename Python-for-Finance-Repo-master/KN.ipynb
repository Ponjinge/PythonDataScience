```python
#sklearn import
import sklearn
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
        
#matplotlib import
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Charger le dataset IRIS
diabetes = datasets.load_diabetes()
       
X = diabetes.data
y = diabetes.target
print("Shape of diabetes dataset = " + str(diabetes.data.shape))
print("Dataset column : " + str(diabetes.feature_names))
       
plt.scatter(X[:,2], y)
```

```python
#On va faire une étude des K Plus Proche Voisins
#We will do a study of the K Nearest Neighbors
```

```python
#On sépare le dataset avec 75% d'entraînement et 25% de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
plt.plot(X_train[:,2], y_train, 'o', label='train')
plt.plot(X_test[:,2], y_test, 'o', label='test')
plt.legend(loc='best')
```

```python
 #Entraînement d'une régression linéaire et évaluation de la régression linéaire
```

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_train = knn.predict(X_train)
#On affiche un nuage de point avec les données train, test et les données prédite
plt.plot(X_train[:,2], y_train, 'o', c="green", label="train", markersize=2)
plt.plot(X_test[:,2], y_test, 'o', c="red", label='test', markersize=2)
plt.plot(X_train[:,2], y_pred_train, 's', c="blue", label="prediction", markersize=3)
plt.legend(loc='best')


#On affiche le score
print("Score : " + str(knn.score(X_test, y_test)))
```

```python
#Les données prédite semblent correspondre aux données train et aux données test sur le graphique. 
#le score est de ~0.45 ce qui signifie que 45% des données sont prédites correctement.
#Le modèle est donc moyennement fiable
```

```python
#L'équation linéaire
```

```python
#Create the objet
regr = linear_model.LinearRegression()
#Train it 
regr.fit(X_train, y_train)

# Predictions
diabetes_y_pred = regr.predict(X_test)

#Display

print("Coefficient : " + str(regr.coef_))
print("Coordonnée à l'origine : " + str(regr.intercept_))
print("Equation : ")
print("Y =", end="")

#Display the coefficient for each column
x = 0
for i in diabetes['feature_names']: 
    print(  str(regr.coef_[x])+ "*"+ i   + " + ", end="")
    x = x+1
    print(regr.intercept_)
```
