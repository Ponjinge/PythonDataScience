```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
#Funtional
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object orientated
fig = plt.figure()

axes= fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('y label')
axes.set_title('tit')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title('Smallplot')
axes1.set_title('Largeplot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title("tit")
axes[1].plot(y,x)
axes[1].set_title("tirt")
```

```python
fig=plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig= plt.figure()

ax= fig.add_axes([0,0,1,1])
ax.plot(x,x**2+4*x+5, '--r', label='poly')
ax.plot(x,x**2, '*-g', label= '')
ax.set_title('eeee')
ax.set_ylabel('ylabel')
ax.set_xlabel('this the x label')
ax.legend(loc=0)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2+x*4, color='purple', lw=0.9, ls='--', marker='o', markersize=10, 
       markerfacecolor='y', markeredgewidth=3,markeredgecolor='g')
ax.set_xlim([1,3])
ax.set_ylim([5,20])
```
